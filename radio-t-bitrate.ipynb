```python
import pandas as pd
import os
import json
import matplotlib.pyplot as plt
from natsort import natsorted, ns

df = pd.DataFrame(columns=['Number', 'Length', 'Size'])

for filename in natsorted(os.listdir('data/episodes/')):
    if filename.endswith(".json"): 
        with open('data/episodes/' + filename) as f:
            data = json.load(f)
            df = df.append({'Number': data['number'], 'Length' : data['file']['length_seconds'],'Size' : data['file']['size_bytes']}, ignore_index=True)
            
plt.rcParams["figure.figsize"] = (20,10)

plt.xlabel('episode number')
plt.ylabel('Size / Length')

plt.plot(df['Number'], df['Size']/df['Length'])

plt.xlim(xmin=0)
plt.ylim(ymin=0)

plt.show()            
```
